```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

# 1. Load the environment variables
load_dotenv()

# 2. Use 'gemini-flash-latest' which is available for your key
llm = ChatGoogleGenerativeAI(
    model="models/gemini-flash-latest",
    google_api_key=os.getenv("GOOGLE_API_KEY")
)

print(llm.invoke("eassay on ai").content)
```

```python
llm.invoke("eassay on ai").content
```

```python
for chunk in llm.stream("eassay on ai"):
    print(chunk.content)
```

```python
### batch
response= llm.batch(
    ["eassay on ai","eassay on malav"],
    config={"max_concurrency":2}           ) 

print(response)
```
